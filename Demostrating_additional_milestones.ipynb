```python
#using Numpy function to add a new column for each dataframe before merging them

data_2010['Year'] = np.repeat(2010,data_2010.shape[0])
```

```python
#before merging df I will add a new column to all df created to show the Year only 

data_2011.insert(loc=0, column='Year', value = 2011)
data_2012.insert(loc=0, column='Year', value = 2012)
data_2013.insert(loc=0, column='Year', value = 2013)
data_2014.insert(loc=0, column='Year', value = 2014)
data_2015.insert(loc=0, column='Year', value = 2015)
data_2016.insert(loc=0, column='Year', value = 2016)
data_2017.insert(loc=0, column='Year', value = 2017)
data_2018.insert(loc=0, column='Year', value = 2018)
data_2019.insert(loc=0, column='Year', value = 2019)
data_2020.insert(loc=0, column='Year', value = 2020)
data_2021.insert(loc=0, column='Year', value = 2021)

#showing head of first df to check new column

data_2010.head(-1)
```

# Demonstating merging dataframe

```python
#merging all dataframes from years 2017 to 2021 with contact fuction

housing_data = pd.concat([data_2010,
                          data_2011,
                          data_2012,
                          data_2013,
                          data_2014,
                          data_2015,
                          data_2016,
                          data_2017,
                          data_2018, 
                          data_2019,
                          data_2020, 
                          data_2021],
                         axis=0)

housing_data.head(-1)
```

```python
#checking missing values

housing_data.isnull() #missing data in some columns
```

```python
#checking if there are any missing values in Price column

print(housing_data['Price (€)'].isnull().values.any()) #no missing data in Price 
```

# Demonstarting replacing missing values 

```python
#demostarting replacing missing values 

housing_data['Postal Code'] = housing_data['Postal Code'].fillna('Empty')

housing_data.head()
```

```python
#cheking for duplicates 

print(housing_data.Eircode.value_counts())

housing_data['Eircode'].duplicated().any()

housing_data['Eircode'].is_unique
```

```python
#dropping duplicates and keeping only the last occurance

housing_data['Eircode'].drop_duplicates(keep='last')
```

```python
housing_data['Price (€)'] = housing_data['Price (€)'].str.split(',').str.join('').astype(float)

housing_data.dtypes
```

```python
housing_data.sort_values(by=['Price (€)'], ascending=False)
```

```python
#removing scientific annotation from Price column

pd.set_option('display.float_format', lambda x: '%.0f' % x)

housing_data.head(-1)
```

# Demonstrating use of conditional statements

```python
#using for loop and conditional statements to add a new column based on buyer assumptions 
#(Price over 5 million bought by Investor)

result = []

for value in housing_data['Price (€)'] : 
    if value >= 5000000 :
        result.append('Investor')
    else : 
        result.append('Private')
        
housing_data['Buyer'] = result

housing_data.head(-1)
```
